```python
# setup_check.ipynb

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Print shape
print("Iris data shape:", X.shape)   # (150, 4)
print("Iris target shape:", y.shape) # (150,)

# Convert ke DataFrame untuk visualisasi
df = pd.DataFrame(X, columns=iris.feature_names)
df["species"] = y

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="species",
    palette="deep"
)
plt.title("Iris Dataset Visualization")
plt.show()
```
